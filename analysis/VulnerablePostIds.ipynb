```python
import os,csv
import xml.etree.ElementTree as ET
```

```python
def parse(xmlFile, outputTxt):
    print("Parsing " + xmlFile)
    unique_ids = set()
    xml = ET.parse(xmlFile)
    root = xml.getroot()
    errors = root[1]
    for i in errors:
        key = i.get("id")
        print(key, i.get("cwe"))
        if i.get("cwe") is not None and i.get("cwe") != "563":
            loc=i[0]
            fileName = loc.get("file")
            temp = fileName.split("-")
            postId = temp[1].split(".")
            line = postId[0]
            unique_ids.add(line)

    print("total vul posts: ", len(unique_ids))
    outputFile = open(outputTxt,"a")
    for i in unique_ids:
        print(i)
        outputFile.write(i + "\n")
    outputFile.close()
```

```python
# parse("raspberrypi/output-new.xml","raspberrypi/VulIds.txt")
parse("arduino/output-new.xml","arduino/VulIds.txt")
# parse("stackoverflow/output-new.xml","stackoverflow/VulIds.txt")
```

```python
def getUserIds(idsFile, outputTxt, postInfo):
    temp = []
    with open(idsFile) as f:
        temp = f.readlines()
        
    lines = []
    for l in temp:
        lines.append(l.strip())
    
    users = []
    with open(postInfo,encoding="utf8") as f:
        PostInfo = csv.DictReader(f)
        for row in PostInfo:
            if row['PostId'] in lines:
                users.append(row['OwnerUserId'])
    
    outputFile = open(outputTxt,"a")
    for i in users:
        outputFile.write(i + "\n")
    outputFile.close()
```

```python
getUserIds("raspberrypi/rpVulIds1.txt","raspberrypi/rpUsers.txt", "raspberrypi/Post_Information.csv")
getUserIds("arduino/ardVulIds.txt","arduino/ardUsers.txt", "arduino/Post_Information.csv")
getUserIds("stackoverflow/so-VulIds.txt","stackoverflow/soUsers.txt", "stackoverflow/Post_Information.csv")
```
