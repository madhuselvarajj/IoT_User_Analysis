```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import os, csv
```

```python
def replace(text):
    to_omit = ['earlier', 'versions', 'tensorflow','version']
    for o in to_omit:
        text = text.replace(o,'')
    return text
```

```python
text = "WatchGuardFirebox Donglify Snapdragon_Consumer_IOT Snapdragon_Industrial_IOT"\
" Eclipse_IoT Codesys Profinet Snapdragon_Consumer_IOT Snapdragon_Industrial_IOT, GoPro"\
"Cisco_Video_Surveillance_8000_Series_IP_Cameras Snapdragon_Consumer_IOT Snapdragon_Industrial_IOT"\
"SIMATIC_HMI_Comfort_Outdoor_Panels"

word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
word_cloud.to_file("wordcloud.png")
```
